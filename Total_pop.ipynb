```python
#dependices
import csv
import pandas as pd
import os
```

```python
#set path for file
full_table = "Resources/population_by_country_2020.csv"
```

```python
#read csv
file = pd.read_csv(full_table, encoding="utf-8")
file.head()
```

```python
#drop and missing data
clean_table = file.dropna(how="any")
```

```python
new_index = clean_table.set_index("Country (or dependency)")
new_index.head()
```

```python
#Drop unneeded columns (Land Area, Migrants, Fert. Rate, Med. Age, Urban Pop %, World Share)
new_table = new_index.drop(['Yearly Change','Net Change','Land Area (Km²)','Migrants (net)', 'Fert. Rate', 'Med. Age', 'Urban Pop %', 'World Share'], axis=1)
new_table
```

```python
# The columns that were taken off are deemed as not useful. 
# If there is a use for them the orginal dataset can be used to join the tables 
```

```python
# write new csv
new_table.to_csv("Resources/Population_clean.csv", index=False, header=True)
```
