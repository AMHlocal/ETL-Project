```python
# Dependencies and Setup.
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# CSV Path
data = "Resources/time_series_covid19_confirmed_global.csv"
```

```python
# Creating a variable for the DF.
covid_df = pd.read_csv(data)
```

```python
# Looking at the data to see what needs to go. I notice that numbers posted under each date are a rolling count, 
# therefore all I need is the most recent date (9/14/20) to get the current number of cases per country.
covid_df.head(10)
```

```python
# Deleting columns I don't need.
essentials_df = covid_df[["Country/Region", "9/14/20"]]
essentials_df.head(20)
```

```python
# The problem we have above is the duplicate countries which are divided by state/province. We need those combined
# and summed together. 
summed_df = essentials_df.groupby(['Country/Region']).sum()
summed_df.head(20)
```

```python
# Renaming columns to reflect the data
clean_df = summed_df.rename(columns={'9/14/20': "Total COVID-19 Cases (9/14/20)"})
clean_df.head(30)
```

```python
# Pushing the cleaned DF to a CSV file
clean_df.to_csv("Resources/cleaned_covid_cases.csv")
```
