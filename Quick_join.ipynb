```python
import pandas as pd
import os
import csv
```

```python
#population data
pop = "Resources/Population_clean.csv"
covid = "Resources/cleaned_covid_cases.csv"
```

```python
pop_read = pd.read_csv(pop, encoding="utf-8")
pop_read.head()
```

```python
pop_read_2 = pop_read.rename(columns={'Country (or dependency)': 'Country/Region'})
pop_read_2.head()
```

```python
covid_read = pd.read_csv(covid, encoding="utf-8")
covid_read.head()
```

```python
merged_data = pd.merge(pop_read_2, covid_read, how="outer", on="Country/Region")
merged_data
```

```python
#dropping everything to only get a country list
country = merged_data.drop(['Population (2020)', 'Density (P/Km²)', 'Total COVID-19 Cases (9/14/20)'], axis = 1)
country
```

```python
# write new csv
country.to_csv("Resources/master_country.csv", index=False, header=True)
```
